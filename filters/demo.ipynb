```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
# Get x values of the sine wave
time = np.arange(0, 10, 0.1)
amplitude = 5*np.sin(time)
noise = np.random.normal(0,1,100)
noise_signal = amplitude+noise
plt.plot(time, amplitude, label="raw")
plt.plot(time, noise_signal, label="noise")
plt.legend()
```

```python
b, a = signal.butter(10, 0.1, btype='lowpass', analog=False)
low_passed = signal.filtfilt(b, a, noise_signal)
plt.plot(time, low_passed, label="filter")
plt.plot(time, noise_signal, label="input")
plt.plot(time, amplitude, label="raw")

plt.legend()
```
