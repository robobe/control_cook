```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
noise=np.random.normal(0,1,1000)
print(np.shape(noise))
plt.plot(noise)
```

```python
signal = np.full((1,1000), 30)
signal = np.squeeze(signal)+noise
plt.plot(signal)
```

```python
def g_filter(signal, alpha):
    prediction = 0
    estimate = 0
    out = np.zeros((len(signal),1))
    for i in range(len(signal)):
        prediction = estimate
        estimate=prediction+alpha*(signal[i]-prediction)
        out[i] = estimate
    return out
```

```python
out = g_filter(signal, 0.02)
plt.plot(out)
```

```python
def l_filter(signal, alpha):
    out = np.zeros((len(signal), 1))
    history = 0
    beta = 1-alpha
    for i in range(len(signal)):
        result = history*beta + signal[i]*alpha
        out[i] = result
        history = result
    return out
```

```python
out = l_filter(signal, 0.1)
plt.plot(out)
```
